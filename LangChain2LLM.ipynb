```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate

# 🔐 Set your actual Gemini API Key here
os.environ["GOOGLE_API_KEY"] = "AIzaSyDOltwimV1w1RbysIaaiHzAX4-blwxbiew"

# 🔮 Load two Gemini LLMs (same model with different temperatures or same, up to you)
llm_1 = ChatGoogleGenerativeAI(model="gemini-1.5-pro-latest", temperature=0.4)
llm_2 = ChatGoogleGenerativeAI(model="gemini-1.5-pro-latest", temperature=0.3)

# 📜 Prompt to extract core information
extract_prompt = PromptTemplate.from_template("""
You are a smart AI. Read the article below and extract only the key points and important facts:

{raw_article}
""")

# 📜 Prompt to summarize the extracted info
summarize_prompt = PromptTemplate.from_template("""
Summarize the following extracted content into 3–5 concise sentences:

{extracted}
""")

# ⚙️ Combine steps using LangChain-style chaining
# Step 1: Extract → Step 2: Summarize
chain = (
    extract_prompt | llm_1 | (lambda output: {"extracted": output}) | summarize_prompt | llm_2
)

# ✅ Sample news article input
article = """
India has become the world's most populous country, surpassing China, according to the United Nations.
India’s population is estimated to be over 1.428 billion, compared to China’s 1.425 billion.
This marks the first population decline in China in 60 years, while India’s growth, though slowing, continues steadily.
Experts say this shift could lead to significant changes in global economics and politics.
"""

# 🚀 Run the full pipeline
result = chain.invoke({"raw_article": article})
print("🧠 AI Summary:\n", result)
```

```python
import os
import gradio as gr
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain_google_genai import ChatGoogleGenerativeAI

# 🔐 Set Gemini API key
os.environ["GOOGLE_API_KEY"] = "AIzaSyBkWa3LW1TWulg2eqmNcp3g0oGQMYf4Mb0"  # Replace this with your real Gemini API key

# 🎯 Load Gemini models
summarizer_llm = ChatGoogleGenerativeAI(model="gemini-1.5-pro-latest", temperature=0.3)
translator_llm = ChatGoogleGenerativeAI(model="gemini-1.5-pro-latest", temperature=0.3)

# 📝 Prompt for summarization
summarization_prompt = PromptTemplate(
    input_variables=["text"],
    template="""
Summarize the following article in a clear, concise paragraph:

{text}
"""
)
summarizer_chain = LLMChain(llm=summarizer_llm, prompt=summarization_prompt)

# 🌐 Prompt for translation
translation_prompt = PromptTemplate(
    input_variables=["text", "language"],
    template="""
Translate the following text into {language}:

{text}
"""
)
translator_chain = LLMChain(llm=translator_llm, prompt=translation_prompt)

# ⚙️ Functions
def summarize_text(text):
    try:
        return summarizer_chain.run(text=text)
    except Exception as e:
        return f"⚠️ Error during summarization: {str(e)}"

def translate_text(text, language):
    try:
        return translator_chain.run(text=text, language=language)
    except Exception as e:
        return f"⚠️ Error during translation: {str(e)}"

# 🎨 Gradio UI
with gr.Blocks(title="Gemini 1.5 Summarizer + Translator") as demo:
    gr.Markdown("# 🧠 AI Summarizer & 🌍 Translator with Gemini 1.5")

    with gr.Tab("📝 Summarizer"):
        text_input = gr.Textbox(lines=6, label="Enter article/text to summarize")
        summarize_button = gr.Button("Summarize")
        summary_output = gr.Textbox(label="Summary")

        summarize_button.click(fn=summarize_text, inputs=text_input, outputs=summary_output)

    with gr.Tab("🌐 Translator"):
        translation_text = gr.Textbox(lines=4, label="Text to Translate")
        language_input = gr.Textbox(label="Target Language (e.g., French, Hindi, Spanish)")
        translate_button = gr.Button("Translate")
        translated_output = gr.Textbox(label="Translated Text")

        translate_button.click(fn=translate_text, inputs=[translation_text, language_input], outputs=translated_output)

demo.launch(debug=True)
```
