```python
!pip install transformers gradio
from transformers import pipeline
import gradio as gr

# Load the summarization pipeline from Hugging Face
summarizer = pipeline("summarization")

def summarize_text(text):
    """Summarizes input text using a Hugging Face summarization model."""
    if not text:
        return "Please enter some text to summarize."

    try:
        summary = summarizer(text, max_length=150, min_length=30, do_sample=False) #Adjust params as needed.
        return summary[0]['summary_text']
    except Exception as e:
        return f"An error occurred: {e}"

# Gradio interface
iface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(lines=10, placeholder="Enter text to summarize..."),
    outputs=gr.Textbox(lines=5, placeholder="Summary will appear here..."),
    title="Hugging Face Text Summarizer",
    description="Enter text and a Hugging Face model will generate a concise summary.",
)

iface.launch()
```
